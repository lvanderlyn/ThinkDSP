Problem 1 - **Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?**

```python
def dft(ys):
    N = len(ys)
    ts = np.arange(N) / N
    freqs = np.arange(N)
    args = np.outer(ts, freqs)
    M = np.exp(i * 2*np.pi * args)
    amps = M.conj().transpose().dot(ys)
    return amps

def idft(amps):
    ys = dft(amps) / N
    return ys

def synthesize2(amps, freqs, ts):
    args = np.outer(ts, freqs)
    M = np.exp(i * 2*np.pi * args)
    ys = np.dot(M, amps)
    return ys

def analyze2(ys, freqs, ts):
    args = np.outer(ts, freqs)
    M = np.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys) / N
    return amps
```

```python
N = 4
amps = np.array([0.6, 0.25, 0.1, 0.05])

ys = idft(amps)
thinkplot.plot(ys)
print(dft(ys))
```

**Comments: ** Although the range of frequencies has changed, the effect is not obvious from the printout they give. It turns out that the dft is by necessity a periodic function and so shifting the range of frequencies will merely shift which part of the periodic function is observed. When we go from N to N-1, the effect is basically the same, due to aliasing and the periodic nature of the dft. When we shift by half of the length of ys, it's a little bit more difficult to see that it is the same function we are looking at, however, if one reads the printed values, it becomes clear that they are merely start over by 2 values or one could say they have been shifted by half the length of ys. However, it is also important to understand that although the real parts of the signal are the same among all of the shifted values, the imaginary, or phase part of the signal has definitely changed to denote the shift

Problem 3 - ** So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.**

```python
signal = thinkdsp.SquareSignal(freq=440)
wave = signal.make_wave(duration=1, framerate=44100)
window = np.ones(11)
window /= sum(window)
thinkplot.plot(window)
segment = wave.segment(duration=0.01)
thinkplot.config(xlabel='time', ylabel='', title='Boxcar Window')
```

```python
def zero_pad(array, n):
    res = np.zeros(n)
    res[:len(array)] = array
    return res

N = len(segment)
boxcar = zero_pad(window, N)
thinkplot.plot(boxcar)
thinkplot.config(xlabel='time', ylabel='', title='Boxcar Window')
```

```python
spec = thinkdsp.Wave(boxcar, 11025).make_spectrum()
spec.plot()
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', title='Effect of Boxcar Filter')
```

```python
thinkplot.plot(spec.fs, np.angle(spec.hs))
thinkplot.config(xlabel='frequency (Hz)', ylabel='Pahse offset', title='Effect of Boxcar Filter')
```

```python
gaussian = scipy.signal.gaussian(M=11, std=2)
gaussian /= sum(gaussian)
thinkplot.plot(gaussian)
thinkplot.config(xlabel='time', ylabel='', title='Gaussian Window')
```

```python
spec2 = thinkdsp.Wave(gaussian, 11025).make_spectrum()
spec2.plot()
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', title='Effect of Gaussian Filter')
```

```python
thinkplot.plot(spec2.fs, np.angle(spec2.hs))
thinkplot.config(xlabel='frequency (Hz)', ylabel='Phase Offset', title='Effect of Gaussian Window')
```

**Comments: ** In this problem, I used code from Allen Downey's thinksdsp Chap08.ipynb to create the boxcar and Gaussian windows. Once I had these windows, I simply converted them into the frequency domain and used np.angle() to find the phase offset from the hs of their spectra. After doing this I found that for the boxcar, the offset seemed to be somewhat periodic, but was definitely increasing in magnetude over time. The Gaussian, in comparison had less components, but seemed to be somewhat less periodic as well, with a sharp change in amplitude after 3000 Hz.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f. If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like. Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?**
