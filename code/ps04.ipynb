Can you construct a vector of complex amplitudes that shifts all components by the same amount of time?**

```python
class ComplexSinusoid(thinkdsp.Sinusoid):
    """Represents a complex exponential signal."""
    
    def evaluate(self, ts):
        """Evaluates the signal at the given times.

        ts: float array of times
        
        returns: float wave array
        """
        print(ts)
        i = complex(0, 1)
        phases = PI2 * self.freq * ts + self.offset
        print(phases)
        ys = self.amp * numpy.exp(i * phases)
        return ys
```

```python
PI2 = 2 * math.pi
i = complex(0, 1)
amps = np.array([0.6, 0.25, 0.1, 0.05])
freqs = [100, 200, 300, 400]
framerate = 11025

ts = np.linspace(0, 1, framerate)

def synthesize2(amps, freqs, ts):
    args = np.outer(ts, freqs)
    M = np.exp(i * PI2 * args)
    ys = np.dot(M, amps)
    return ys
```

Problem 2 - **In Chapter 7, we represented the DFT as matrix multiplication, but the more common presentation is as a summation: From http://en.wikipedia.org/wiki/Discrete_Fourier_transform.
```python
def my_dft(ys):
    """
    Takes in a list of ys and uses that to determine the frequencies contained within the signal.
    """
    hs = []
    for k in range(len(ys)-1):
        x_k = 0
        for n in range(len(ys)):
            x_k += ys[n]*np.exp(-1j*np.pi*2*k*n/len(ys))
        hs.append(x_k)
    return hs

hs = my_dft(ys2[0:1000])
# print(hs)
spec_test = thinkdsp.Spectrum(hs, framerate)
spec_test.plot()
thinkplot.config(xlim=[0,2000],title = 'Spectrum using my dft implemntation', xlabel = 'frequency (Hz)', ylabel = 'amplitude')
```

Problem 3 - **In Chapter 7, I showed how we can express the DFT and inverse DFT as matrix multiplications. These operations are relatively fast, taking time proportional to N2, where N is the length of the wave array. That would be fast enough for many applications, but it turns out that there is a faster algorithm, the Fast Fourier Transform (FFT), which takes time proportional to N logN. Read about the FFT at http://en.wikipedia.org/wiki/Cooley-Tukey_FFT_algorithm, and write an implementation. Hint: I suggest you write a simple version as a recursive function; don’t worry about “data reordering, bit reversal, and in-place algorithms”.**

```python
def wikipediaFFT(ys):
    """
    computes the fast fourier transform on a set of ys, as long as the length is a power of two
    """
    N = len(ys)
    if bin(N).count("1") != 1:
        print('not power of two')
        return

    W = np.exp(-1j * PI2 * np.arange(N) / N)
    if N == 1: 
        return ys
    else:
        evens = wikipediaFFT(ys[::2])
        odds = wikipediaFFT(ys[1::2])
        x0 = evens + W[:N/2] * odds
        x1 = evens + W[N/2:] * odds
        return np.concatenate([x0,x1])
```

Problem 4 - ** Using the timing code from PS03, compare the performance of my implementation of autocorr from Chapter 5, my fft_autocorr from Chapter 8, and numpy.correlate.  Can you tell whether NumPy is using the FFT algorithm?**

```python
def corrcoef(xs, ys):
    return np.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
def zero_pad(array, n):
    res = np.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = np.fft.fft(signal)
    fft_window = np.fft.fft(window)
    return np.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=1.0)
ns = 2 ** np.arange(4, 12)

results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o autocorr(thinkdsp.Wave(ys, framerate))
    results.append(result)
bests = [result.best for result in results]
```

```python
results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o fft_autocorr(ys)
    results.append(result)
bests2 = [result.best for result in results]
```

```python
results = []
for N in ns:
    ys = noise.ys[:N]
    result = %timeit -o np.correlate(ys, ys, 'same')
    results.append(result)
bests3 = [result.best for result in results]
```

```python
pyplot.plot(ns, bests, label="autocorr")
pyplot.plot(ns, bests2, label="fft_autocorr")
pyplot.plot(ns, bests3, label="np.correlate")
pyplot.legend()
thinkplot.config(title='Comparing efficency of algorithms', xscale='log', yscale='log', loc=2)
pyplot.show()
```

**Comments: ** The comparison of the algorithms shows that the np.correlate function is much more efficient for smaller numbers, but becomes rapidly less effecient as time passes, so if one were to need to correlate a very long signal, it would be more efficient to use Allen Downey's fft_autocorr function.
