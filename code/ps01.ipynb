Problem 5 -- **Synthesize a wave by creating a spectrum with arbitrary harmonics, invoking make_wave, and listening. What happens as you add frequency components that are not multiples of the fundamental?** 

```python
base = np.random.random_integers(0,500)
n = 10
mix = 0
for i in range(n):
    cos_sig = thinkdsp.CosSignal(freq= base*i, amp=5./(i+1), offset=0)
    mix = mix + cos_sig
wave = mix.make_wave(duration=0.5, start=0, framerate=11025)
spec = wave.make_spectrum()
spec.plot()
wave.make_audio()
```

**Comments: ** When I add frequencies that are not multiples of the fundamental frequency the sound gains a dissonant sound. It is no longer as pleasant to listen to, nor is it as clear as when there are only multiples of the dominant frequency

Problem 6 -- **A sawtooth signal has a waveform that ramps up linearly from -1 to 1, then drops to -1 and repeats**

```python
#Create a sawtooth signal based on existing signal class

class Sawtooth(thinkdsp.Signal):
    def __init__(self, frequency=450, offset=0):
        """
        initializes classfrequency which
        
        inputs:
            frequency -- determines the slope of the sawtooth
            offset -- determines where on the sawtooth the time 0 is located
        """
        self.frequency = frequency
        self.offset = offset
          
    def evaluate(self, ts):
        
        cycles = (self.frequency * ts + self.offset) / 2*np.pi
        frac = cycles % (2*np.pi)
        centered = frac- np.mean(frac)
        return centered/float(max(centered))

    
test = Sawtooth(50)
test.plot()
# ts = np.array([i* 0.01 for i in range(1000)])
# ys = test.evaluate(ts)

    
```
