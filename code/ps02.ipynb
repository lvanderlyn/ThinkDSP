Problem 2 -- **Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly.**

```python
class SawToothChirp(thinkdsp.Chirp):
    def _evaluate(self,ts,freqs):
        dts = np.diff(ts)
        dps = math.pi * 2 * freqs * dts
        phases = np.cumsum(dps)
        phases = np.insert(phases, 0, 0)
        cycles = phases
        frac, _ = np.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac),self.amp)
        return ys
```

```python
test = SawToothChirp(start=220, end=880)
test_wave = test.make_wave(duration=1)
test_wave.segment(start=0, duration=0.01).plot()
thinkplot.config(xlabel='time (s)', ylabel='amplitude', legend=False)
test_wave.make_audio()
```

```python
test_spec = test_wave.segment(start=0, duration=0.01).make_spectrum()
test_spec.plot()
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

```python
spect = test_wave.make_spectrogram(seg_length=600)
spect.plot()
thinkplot.config(xlabel='time(s)', ylabel='frequency(Hz)', legend=False)
```

**Comments: ** The above spectrogram shows that the dominant frequency is increasing linearily with time (the dark blue line) which can be difficult to tell for sure when looking only at the spectrum or the plot of the original wave. The effect of aliasing can be clearly seen in the triangle waves shown behind it. Rather than continuing to increase linearly, these waves fold at the frequency which represents half of the sampling rate, in this example that is around 5500 Hz, they fold again at zero which is what causes some of the forms in the spectrogram to look like they are bouncing. As for why there are multiple waves, I believe this is a result of the keeping the harmonic spacing constant. The dominant frequency is increasing linearly, but if all of the frequencies were just linearly shifted, they would no longer be multiples of the fundamental. In order to preserve this spacing, they need to increase linearly at different rates, which can be seen by the varying slopes of the triangle waves. 

Problem 3 - **In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length. Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp? Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.**


```python
def gliss(d_len, A=1,start_freq=262, end_freq=349, bounce=False):
    """
        Takes in the starting and ending frequencies and returns a wave object representing the glissando between them
        
        # Formulae for converting b/n pitch and frequency come from: 
        # http://people.sju.edu/~rhall/SoundingNumber/pitch_and_frequency.pdf
        
        inputs:
            d_len -- the speed in cm/s the trombone length of changing
            A -- the constant of proportionality between length and pitch
            start -- the starting frequency of the glissando
            end -- the ending frequency of the glissando
            bounce -- boolean of whether the gliss is repeated from end_freq to start_freq
            
        returns:
            gliss_wave -- a waveform object representing the glissando
    """
   
    #Converting start/stop frequencies to pitches for easier manipulation
    start_pitch = 69+ (12*np.log2(start_freq/440.0))
    end_pitch = 69+ (12*np.log2(end_freq/440.0))
    
    #calculated based on 7 known trombone length to note names (from google)
    start_len = 8.3 * start_pitch - 2.65
    end_len = 8.3 * end_pitch - 2.65
    
    #Establishing time range/time spacing b/n frequencies
    time = (end_len-start_len)/float(d_len)
    d_pitch = (end_pitch - start_pitch)/float(time)
    frame_rate = 22050
    dt = frame_rate * time
    ts = np.linspace(0, time, dt)
    
    #Problem states that the pitch is inversely proportional to the length of the trombone and len changes at constant rate
    pitches = start_pitch + d_pitch * ts
    
    #converting back to frequency with formula from above website
    freqs = 440 * 2**((pitches-69)/12.0)
    freqs = np.append(freqs, freqs[::-1])
    ts = np.append(ts, ts+dt)
    
    #Making it into a wave
    dts = np.diff(ts)
    dps = 2*np.pi * freqs * (np.append(np.array([0]), dts))
    phases = np.cumsum(dps)
    phases = np.insert(phases, 0, 0)
    ys = np.cos(phases)
    glis_wave = thinkdsp.Wave(ys, 11025)
    return glis_wave

gliss(20).make_audio()
    
```

**Comments: ** In this problem, I created a wave which had a frequency based on a linearly increasing pitch. Originally, I had a calculation relating the length of the trombone to the pitch it produced, however after much work to get this to line up correctly for all of the conversions, I realized that because there was in fact a linear relationship between pitch and trombone length (as found from internet data and plotting the trendline) it was simpler to determine the start and end length of the trombone needed to produce the start and end frequencies a user would input into the function. From this I assumed that the motion of the slide was completely linear and used a user-inputted rate of change of slide to determine the total amount of time needed to transition between notes. Based on the fact that the length of the trombone was linearly proportional to the pitch produced, I increased the pitch at a constant rate over the time period and then converted the pitch values back to frequency values. In order to have the full up and down effect of the gliss which the problem asked for, I appended the reversed gliss to my np.array of frequncies.  


Once I had a list of frequencies, I was able to use a code very similar to the evalueate code from Signal. This allowed me to generate a wave form which had a pitch that varied linearly from the start to stop frequency. By extension, this means that the frequency varied in an exponential manner.

Problem 6 - **In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.
To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment of bitcoin price data, and then compute the average across the segments. You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method. **

```python
ys = []
with open('coindesk-bpi-USD-close_data-2015-01-31_2015-02-01.csv', 'rb') as csvfile:
    bit_read = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in bit_read:
        if len(row) > 1:
            data = row[1]
            ys.append(data)
ys = np.array(ys[1::])
ts = np.linspace(0, len(ys), len(ys))
bit_wave = thinkdsp.Wave(ys, 2/60.0)
bit_wave.plot()
thinkplot.config(xlabel='time since opening (s)', ylabel='price (USD)', legend=False)
```

```python
steps = 5
step_len = len(bit_wave.ys)/float(steps)
spec = bit_wave.segment(start=0, duration=step_len).make_spectrum()
power_sum = spec.amps ** 2
freqs = spec.fs
for step in range(1, steps):
    slice_spec = bit_wave.segment(start=step-1, duration=step_len).make_spectrum()
    power_sum += slice_spec.amps ** 2
avg_pwr = power_sum/float(steps)
thinkplot.Plot(freqs, avg_pwr)
thinkplot.show(xlabel='frequency (Hz)', ylabel='average power density', xscale='log', yscale='log')
```

**Comments: ** Using the Bartlett method, I divided the sample up into five groups and then took the average power for each frequency and plotted it in the figure above. This can be compared with the figure below, which I merely plotted. Doing this eliminates a lot of noise from the signal and allows the general trend of the curve to be seen more clearly. In this case, the averaged shape shows that bitcoin data is an example of Brownian noise.
